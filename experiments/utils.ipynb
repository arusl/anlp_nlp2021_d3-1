# Utility Functions

### Function: Train a SentencePiece model

```python
def trainSP(ori_path, sampled_path, vocab_size=32000, num_rows=10000):
  df_train_ori = pd.read_csv(ori_path, header=None)
  df_sampled = df_train_ori.copy()
  df_test = df_sampled[0:num_rows]
  df_test.to_csv(sampled_path, index=False)

  start = time.time()
  spm.SentencePieceTrainer.train(input=sampled_path, model_prefix='sp-model-'+vocab_size+'-'+num_rows, vocab_size=vocab_size)
  end = time.time()
  print(end-start)
```

### Example: Train a SentencePiece model

```python
ori_path = '../data/rakuten-sentiment-dataset/binary/binary_train.csv'
sampled_path = '../data/rakuten-sentiment-dataset/sampled-binary-train.csv'
trainSP(ori_path, sampled_path, 32000, 10000)
```

### Function: Create a Sampled Dataset

```python
def sampleData(label, fraction):
    if label=='binary':
        bin_train = pd.read_csv('../data/rakuten-sentiment-dataset/binary/binary_train.csv', header=None)
        bin_test = pd.read_csv('../data/rakuten-sentiment-dataset/binary/binary_test.csv', header=None)
        print("Number of binary training samples: ", len(bin_train))
        print("Number of binary testing samples: ", len(bin_test))
        sample_train = bin_train.sample(frac=fraction, random_state=0)
        sample_test = bin_test.sample(frac=fraction, random_state=0)
        sample_train.to_csv('../data/rakuten-sentiment-dataset/binary/sampled_binary_train.csv', index=False, header=None)
        sample_test.to_csv('../data/rakuten-sentiment-dataset/binary/sampled_binary_test.csv', index=False, header=None)
    elif label=='full':
        full_train = pd.read_csv('../data/rakuten-sentiment-dataset/full/full_train.csv', header=None)
        full_test = pd.read_csv('../data/rakuten-sentiment-dataset/full/full_test.csv', header=None)
        print("Number of full training samples: ", len(full_train))
        print("Number of full testing samples: ", len(full_test))
        sample_train = full_train.sample(frac=fraction, random_state=0)
        sample_test = full_test.sample(frac=fraction, random_state=0)
        sample_train.to_csv('../data/rakuten-sentiment-dataset/full/sampled_full_train.csv', index=False, header=None)
        sample_test.to_csv('../data/rakuten-sentiment-dataset/full/sampled_full_test.csv', index=False, header=None)
```
